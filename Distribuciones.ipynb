```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips=sns.load_dataset("tips")#carga data set predefinidos por seaborn
```

```python
sns.histplot(data=tips,x='tip',bins=15,cumulative=False,hue='sex',stat='count',multiple='dodge') #histplot crea un Histograma
#stat= probability muesta la probabilidad
#https://seaborn.pydata.org/generated/seaborn.histplot.html --> visitar para ver mas parametros de histplot
```

```python
sns.kdeplot(data=tips,x='tip',hue='sex',cumulative=False)
plt.show()
#https://seaborn.pydata.org/generated/seaborn.kdeplot.html-->revisar para mas paremtros
```

```python
sns.ecdfplot(data=tips,x='tip',hue='sex',stat='count')
#https://seaborn.pydata.org/generated/seaborn.ecdfplot.html -->mas info
```

```python
sns.displot(data=tips,x='tip',hue='sex',kind='kde')
#https://seaborn.pydata.org/generated/seaborn.displot.html-->mas info
#kind modifica el tipo de grafico
```

```python
sns.displot(data=tips,x='tip',hue='day',kind='kde')
```

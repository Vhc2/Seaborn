```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips=sns.load_dataset("tips")
tips
```

```python
#SCATTERPLOT grafico de dispersion--> https://seaborn.pydata.org/generated/seaborn.scatterplot.html
sns.scatterplot(data=tips,x='total_bill',y='tip',style='time',size='size')
plt.legend(loc='right')
```

```python
markers={"Lunch":"D","Dinner":"s"}#especifica que markers usar, dependiendo que valor obtenga
sns.scatterplot(data=tips,x='total_bill',y='tip',style='time',size='size',markers=markers,)
plt.legend(loc='right',bbox_to_anchor=[1.25, 0.5])
```

```python
sns.lineplot(data=tips,x='total_bill',y='tip',style='time',size='size')
plt.legend(loc='right',bbox_to_anchor=[1.25, 0.5])
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',style='time',size='size',col='time')
```

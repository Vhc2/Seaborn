```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips=sns.load_dataset('tips')
tips
```

```python
#https://seaborn.pydata.org/generated/seaborn.countplot.html--> more info
sns.countplot(data=tips,x='day',hue='sex')#hue permite segemetnar por un valor
```

```python
#https://seaborn.pydata.org/generated/seaborn.stripplot.html-->more info
sns.stripplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True)
```

```python
#https://seaborn.pydata.org/generated/seaborn.swarmplot.html --more info
sns.swarmplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True)
```

```python
sns.boxplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True)
sns.swarmplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True,color='0')
#https://seaborn.pydata.org/generated/seaborn.boxplot.html -->more info
```

```python
sns.violinplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True)

#https://seaborn.pydata.org/generated/seaborn.violinplot.html-->more info
```

```python
sns.violinplot(data=tips,y='total_bill')
```

```python
#https://seaborn.pydata.org/generated/seaborn.catplot.html --> catplot
sns.catplot(data=tips,x='day',y='total_bill',hue='sex',col='time')#col, lo separa por la varriable categoria time
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips=sns.load_dataset("tips")
tips
```

```python
#JOINPLOT -->https://seaborn.pydata.org/generated/seaborn.jointplot.html
sns.jointplot(data=tips,x='total_bill',y='tip',hue='sex',kind='hist',marginal_ticks=True
              ,marginal_kws=dict(bins=25,fill=False,multiple='dodge'))#marginal ticks, muestra el valor de los graficos externos
#marginal_kws --> sirve para modificar los graficos exteriores
plt.show()
```

```python
#PAIRPLOT -->https://seaborn.pydata.org/generated/seaborn.pairplot.html
#Sirve para variables numericas
sns.pairplot(data=tips,hue='sex',corner=True)#corner-->evita que se hagan comparacion de las mismas variables
```
